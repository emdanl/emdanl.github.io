# Music

```python
# Upload the CSV File

import pandas as pd

spotify = pd.read_csv("https://bcdanl.github.io/data/spotify_all.csv")
spotify.head()
```

```python
# Frequency

artist_counts = spotify['artist_name'].value_counts()
artist_counts.head(10)
```

```python
# Usher Filter

usher_tracks = spotify[spotify['artist_name'] == 'Usher']
usher_tracks[['playlist_name', 'track_name', 'pos']]
```

```python
# Sort Usher

usher_sorted = usher_tracks.sort_values(by='pos')
usher_sorted[['playlist_name', 'track_name', 'pos']]
```
